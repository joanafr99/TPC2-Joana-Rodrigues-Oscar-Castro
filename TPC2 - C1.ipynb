```python
profitlist = company_sales_dataDF ['total_profit'].tolist()
facecreamlist = company_sales_dataDF ['facecream'].tolist()
facewashlist = company_sales_dataDF ['facewash'].tolist()
toothpastelist = company_sales_dataDF ['toothpaste'].tolist()
bathingsoaplist = company_sales_dataDF ['bathingsoap'].tolist()
shampoolist = company_sales_dataDF ['shampoo'].tolist()
moisturizerlist = company_sales_dataDF ['moisturizer'].tolist()

monthList = company_sales_dataDF ['month_number'].tolist()

x=monthList
y1=facecreamlist
y2=facewashlist
y3=toothpastelist
y4=bathingsoaplist
y5=shampoolist
y6=moisturizerlist

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.set_title("Sales data", fontsize=20)
ax.set_xlabel("Month Number", fontsize= 14)
ax.set_ylabel("Sales units in number", fontsize= 14)

plt.xlim({0,12})
plt.ylim({0,18000})

plt.plot(x, y1, "tab:blue", marker= 'o' , label="Face Cream Sales Data")
plt.plot(x, y2, "tab:orange", marker= 'o', label="Face Wash Sales Data" )
plt.plot(x, y3, "tab:green", marker= 'o', label="Toothpaste Sales Data")
plt.plot(x, y4, "tab:red", marker= 'o', label="Bathingsoap Sales Data")
plt.plot(x, y5, "tab:purple", marker= 'o', label="Shampoo Sales Data" )
plt.plot(x, y6, "tab:brown", marker= 'o', label="Moisturizer Sales Data")

plt.legend(loc="upper left")

print("C3. Sales Data")
plt.show()
```

```python
df =  pd.read_csv('company_sales_data.csv')
%matplotlib inline
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(6, 3), subplot_kw=dict(aspect="equal"))

totalunits = df["total_units"].sum()
facecream_data = df["facecream"].sum()
facewash_data = df["facewash"].sum()
toothpaste_data = df["toothpaste"].sum()
bathingsoap_data = df["bathingsoap"].sum()
shampoo_data = df["shampoo"].sum()
moisturizer_data = df["moisturizer"].sum()

facecream = ((facecream_data) / (totalunits)) 
facewash = ((facewash_data) / (totalunits)) 
toothpaste = ((toothpaste_data) / (totalunits))
bathingsoap = ((bathingsoap_data) / (totalunits))
shampoo = ((shampoo_data) / (totalunits))
moisturizer = ((moisturizer_data) / (totalunits))

exp_vals= [facecream, facewash, toothpaste, bathingsoap, shampoo, moisturizer]
exp_labels = ["Facecream", "Facewash", "Toothpaste", "Bathingsoap", "Shampoo", "Moisturizer"]

plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, radius=1, autopct=("%0.1f%%"))
plt.title('Sales Data')

ax.legend(exp_labels,
          title="Sales Data",
          loc="lower right",
          bbox_to_anchor=(0.7, 0.1, 0.5, 0))

plt.show()
```

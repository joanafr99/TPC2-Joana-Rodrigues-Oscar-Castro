### <font color = 'green'>World Cup with Pandas</font>
Solve the problems using the data in Pandas dataframes.

```python
# Read the Players and Teams data into dataframes
f = open('Players.csv')
playersDF = pd.read_csv(f)
f = open('Teams.csv')
teamsDF = pd.read_csv(f)
```

```python
#Qual jogador (print do ultimo nome) de uma equipa com "ia" no nome, 
#que jogou menos de 200 minutos e fez mais de 100 passes?

jogador = playersDF.loc[playersDF["minutes"]<200].loc[playersDF["passes"]>100]
jogador = jogador.iloc[[value for value in range(len(jogador["team"].values)) if "ia" in jogador["team"].values[value]]]
print(jogador)


# FUNDAMENTAÇÃO:
#  Criamos a variável "jogador" indo buscar os valores ao dataframe denominado "playersDF", e usando o ".loc" para localizar dentro da coluna dos "minutes" os que tinham menos de 200 e na coluna dos "passes" o valores accima de 100
#  UtiLizando um array e usando o método ".iloc" procura-mos os nomes das equipas dos jogadores da coluna "team", que tivessem "ia" no seu nome.
#  Printamos o resultado
```

```python
# Qual é o número médio de passes feitos pelos avançados? Pelos meio-campistas?
# Verifique se a resposta especifica qual é qual e não inclua outras
# posições no seu resultado.


pmedios = playersDF.loc[playersDF["position"]=="midfielder"]["passes"].mean()
pavançados = playersDF.loc[playersDF["position"]=="forward"]["passes"].mean()
                                          
print(((str(pmedios)) + " é o número médio de passes feitos pelos jogadores que são meio-campistas")) , 
print(((str(pavançados)) + " é o número médio de passes feitos pelos jogadores que são avançados"))


# FUNDAMENTAÇÃO:
#  Criamos duas variáveis sendo elas o "pmedios" e o "pavançados".
#  Dentro destas variáveis usamos o médoto ".loc" para localizar na coluna "position", as posições respetivas e os seus passes na coluna "passes", e depois usando ".mean()" para calcular a média aritmetica.
#  Printamos o resultado
```

```python
# Qual equipe tem a maior relação (razão, racio, etc) de goalsFor versus goalsAgainst?
# Imprima apenas o nome da equipa.
# Dica: adicione uma coluna "racio" ao dataframe das equipas e, depois, faça o seu "sort" e, 
# em seguida, use a head(1) ou tail(1) ...

teamsDF["racio"] = teamsDF["goalsFor"]/teamsDF["goalsAgainst"]
equipa = teamsDF.sort_values("racio").tail(1)

print("A equipa que tem o maior racio de goalsFor versus goalsAgainst é " + equipa["team"].values[0])


# FUNDAMENTAÇÃO:
#  Para saber qual equipa tinha o maior racio usamos o teamsDF para encontrar os valores "goalsFor" e "golsAgainst" para descobrir a relação.
#  Criamos então a variável "equipa" e usando o método ".sort_values" para sortear os valores através do eixo e usando o ".tail(1)" para buscar o primeiro valor da última linha.
#  Printamos o resultado
```

```python
# Quantos jogadores que jogam numa equipa com ranking <10, 
# jogaram mais de 350 minutos?
# Lembrete: len() fornece o número de linhas que um dataframe contém

equipa = teamsDF.loc[teamsDF["ranking"]<10]["team"].values
jogadores = playersDF.iloc[[team for team in range(len(playersDF["team"])) if playersDF["team"][team] in equipa]]
jogadores = jogadores.loc[jogadores["minutes"]>350]

print( "Os jogadores que jogam numa equipa com ranking <10 e que jogaram mais de 350 minutos foram " + str(len(jogadores)))
jogadores


# FUNDAMENTAÇÃO:
#  Criamos duas variáveis, "equipa" e "jogadores".
#  Para encontrar as equipas com ranking menor que 10, usamos o método ".loc".
#  Para encontrar os jogadores usamos o método ".iloc", para encontra-los dentro da coluna "team" e depois usamos o ".loc" para selecionar os que tinham minutos superiores a 350.
#  Printamos o resultado e usamos o "jogadores" para apresentar uma tabela.
```

### <font color = 'green'>Titanic with Pandas</font>
Solve the problems using the data in Pandas dataframe.

```python
# Read the Titanic data into a dataframe

f = open('Titanic.csv')
titanicDF = pd.read_csv(f)
titanicDF
```

```python
# Quantas mulheres casadas embarcaram em Cherbourg?
# Nota: 'first' é uma função no Pandas, então 'titanic.first' irá gerar um erro;
# use 'titanic[' first ']

casadas = titanicDF.iloc[[value for value in range(len(titanicDF["first"].values)) if "Mrs" in titanicDF["first"].values[value]]]
embkcherbourg = casadas.loc[casadas["embarked"]=="Cherbourg"]

str(len(embkcherbourg))

print("A quantidade de mulheres casadas que embarcaram em Cherbourg é de " + str(len(embkcherbourg)))


# FUNDAMENTAÇÃO:
#  Criamos duas variáveis, "casadas", para defenir as mulheres casadas e "embkcherbourg" para os que embarcaram em Cherbourg.
#  Para o "casadas" usamos o método ".iloc" no dataframe, para encontramos na coluna "first" os nomes "Mrs" e seleciona-los.
#  No "embkcherbourg" usamos o método ".loc" na variável anterior para procurar quais "Mrs" na sua coluna "embarked" tinham Cherbourg.
#  Printamos o resultado usando tambem o "str(len)"" para calcular o número de casadas embarcadas em Cherbourg. 
```

```python
# Liste a tarifa média paga pelos passageiros em cada uma das cidades de embarque.

mediapaga1 = titanicDF.loc[titanicDF["embarked"]=="Southampton"]["fare"].mean()
mediapaga2 = titanicDF.loc[titanicDF["embarked"]=="Queenstown"]["fare"].mean()
mediapaga3 = titanicDF.loc[titanicDF["embarked"]=="Cherbourg"]["fare"].mean()
                                          
print(((str(mediapaga1)) + " é a tarifa média paga pelos passageiros de Southampton")) ,
print(((str(mediapaga2)) + " é a tarifa média paga pelos passageiros de Queenstown")), 
print(((str(mediapaga3)) + " é a tarifa média paga pelos passageiros de Cherbourg"))

# FUNDAMENTAÇÃO:
# Criamos 3 variáveis para as medias pagas por cada cidade.
# Procuramos então pelo valor de "fare" usando o método ".loc" e depois usamos o .mean() para calcular a media.
# Printamos o resultado de cada uma.
```

```python
# Os valores inexistentes (missing)  no Pandas são nulos (null), impressos como 'NaN' e
# a function pd.isnull(v) verifica se um valor é nulo.
# Que fração dos passageiros não tem valor para a idade?

idadeinexistente = titanicDF[titanicDF['age'].isnull()]
idadeinexistente.isnull().sum()

idadeinexistente.age.to_string(index=True)
idadeinexistente.shape[0]

print("A fração dos passageiros que não tem valor para a idade é de " + str(idadeinexistente.shape[0] / titanicDF.shape[0]) )


# FUNDAMENTAÇÃO:
# Criamos a variável "idadeinexistente"
# Dentro desta variável, procuramos dentro da dataframe na coluna "age" quais valores eram impresssos como NaN.
# Usamos o "idadeinexistente.isnull().sum()" para contar os números nulos e utilizamos o "idadeinexistente.age.to_string(index=True)" para verificar todos os NaN e o número corresponde através do "(index=True)" verificando se o número contado estava certo através do "idadeinexistente.shape[0]" 
# Printamos o resultado utilizando a "str()" para calcular a fração do número da "idadeinexistente" pelo número da dataframe.
```
